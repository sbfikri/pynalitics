<p> The data below were exctracted from Kaggle's thread called 'Hotel Booking Demand'. Consisting of 32 columns and more than a thousand rows, the data were gathered to fill the 'scracity of real business data for scientific and educational purpose. </p>

<p> Speaking of purpose, this notebook seeks to identify the likelihood of the hotel bookings get cancelled. The first part of this notebook would explore the nature of this data, that is, its features, missing value, distribution, and outlier. The second part is data tranformation, that includes label encoding, missing value and outlier treatment. The third part of it would create an accurate model to fulfill the very purpose of the notebook </p>

```python
import pandas as pd
import os

directory = os.getcwd()
df = pd.read_csv(directory + '/Practice - 5/hotel_bookings.csv')

df.head()
```

```python
def missing_values():
    data = df.isnull().sum()
    data = pd.DataFrame(data)
    data.reset_index(level = 0, inplace = True)
    data.columns = ['Variable Name', 'Missing Value']
    data = data.sort_values(by = 'Missing Value', ascending = False)
    return data

missing_values()
```

```python
for col in list(df):
    print(col)
    print(df[col].drop_duplicates())
```

```python
df_clean = df.drop(['agent', 
                    'arrival_date_week_number', 
                    'arrival_date_year', 
                    'company', 'previous_bookings_not_canceled', 
                    'reservation_status', 
                    'reservation_status_date',
                    'stays_in_week_nights'
                   ], axis = 1)

df_clean.loc[df_clean['stays_in_weekend_nights'] == 0, 'stays_in_weekend_nights'] = 1
df_clean.loc[df_clean['stays_in_weekend_nights'] > 0, 'stays_in_weekend_nights'] = 0

df_clean.head()
```
